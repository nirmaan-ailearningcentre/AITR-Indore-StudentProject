```python
top_students = df.sort_values(
    by='midterm_score',
    ascending=False
)

top_students.head(5)
```

```python
weak_students = df.sort_values(
    by='midterm_score',
    ascending=True
)

weak_students.head(5)
```

Pass vs Fail Distribution

```python
plt.figure(figsize=(6,4))

sns.countplot(x='pass', data=df)

plt.title('Pass vs Fail Students')
plt.xlabel('Pass Status')
plt.ylabel('Number of Students')

plt.show()
```

```python
plt.figure(figsize=(8,5))

sns.scatterplot(
    x='attendance_pct',
    y='midterm_score',
    hue='pass',
    data=df
)

plt.title('Attendance vs Midterm Score')

plt.show()
```

```python
plt.figure(figsize=(8,5))

sns.scatterplot(
    x='homework_pct',
    y='midterm_score',
    hue='pass',
    data=df
)

plt.title('Homework Score vs Midterm Score')

plt.show()
```

```python
plt.figure(figsize=(8,5))

sns.scatterplot(
    x='study_hours_per_week',
    y='midterm_score',
    hue='pass',
    data=df
)

plt.title('Study Hours vs Midterm Score')

plt.show()
```

```python
top10 = df.sort_values(
    by='midterm_score',
    ascending=False
).head(10)

plt.figure(figsize=(10,5))

sns.barplot(
    x='student_id',
    y='midterm_score',
    data=top10
)

plt.title('Top 10 Students by Midterm Score')

plt.xticks(rotation=45)

plt.show()
```

```python
weak10 = df.sort_values(
    by='midterm_score'
).head(10)

plt.figure(figsize=(10,5))

sns.barplot(
    x='student_id',
    y='midterm_score',
    data=weak10
)

plt.title('Weakest 10 Students')

plt.xticks(rotation=45)

plt.show()
```

```python
plt.figure(figsize=(8,6))

corr = df[['attendance_pct',
           'homework_pct',
           'midterm_score',
           'study_hours_per_week']].corr()

sns.heatmap(
    corr,
    annot=True,
    cmap='coolwarm'
)

plt.title('Correlation Matrix')

plt.show()
```

Insights

INSIGHTS

1. Students with higher attendance percentages generally achieve better midterm scores.

2. Students who complete a larger percentage of homework tend to perform better in examinations.

3. Study hours per week have a positive impact on student academic performance.

4. Most students who passed the exam showed higher attendance, homework completion, and study hours compared to failed students.

5. Attendance, homework performance, and study hours are important indicators of student success.

Conclusion

CONCLUSION
