```python
import pandas as pd
import numpy as np

df = pd.read_csv("Pass-Fail Data.csv")
```

```python
print("First 5 Records:")
print(df.head())
```

```python
print("\nShape of Dataset:")
print(df.shape)
```

```python
print("\nColumns:")
print(df.columns)
```

```python
print("\nData Types:")
print(df.dtypes)
```

```python
print("\nMissing Values:")
print(df.isnull().sum())

print("\nTotal Missing Values:")
print(df.isnull().sum().sum())
```

```python
print("\nStatistical Summary:")
print(df.describe())
```

```python
pass_count = (df['pass'] == 1).sum()
fail_count = (df['pass'] == 0).sum()

print("\nPass Students:", pass_count)
print("Fail Students:", fail_count)
```

```python
pass_students = df[df['pass'] == 1]
fail_students = df[df['pass'] == 0]

print("\nPass Students:")
print(pass_students)

print("\nFail Students:")
print(fail_students)
```

```python
sorted_data = df.sort_values(by='midterm_score', ascending=False)

print("\nTop 10 Students by Midterm Score:")
print(sorted_data.head(10))
```

```python
print("\nAverage Attendance:")
print(np.mean(df['attendance_pct']))

print("\nAverage Homework Percentage:")
print(np.mean(df['homework_pct']))

print("\nAverage Midterm Score:")
print(np.mean(df['midterm_score']))

print("\nAverage Study Hours:")
print(np.mean(df['study_hours_per_week']))
```

```python
print("\nMaximum Midterm Score:")
print(np.max(df['midterm_score']))

print("\nMinimum Midterm Score:")
print(np.min(df['midterm_score']))
```

```python
print("\nCorrelation Matrix:")
print(df.corr(numeric_only=True))
```

```python
print("\nMedian Midterm Score:")
print(np.median(df['midterm_score']))
```

```python
print("\nStandard Deviation of Midterm Score:")
print(np.std(df['midterm_score']))
```
