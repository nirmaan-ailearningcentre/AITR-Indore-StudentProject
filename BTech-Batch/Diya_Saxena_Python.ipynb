# Hospital Patient Records Analysis

This notebook has been fixed to safely load the dataset and avoid common file path errors.

```
import pandas as pd
from pathlib import Path

# Update the CSV file name if needed
csv_file = 'Hospital_patient_records_analysis.csv'

# Check if the CSV file exists
if Path(csv_file).exists():
    df = pd.read_csv(csv_file)
    print("Dataset loaded successfully!")
    display(df.head())
else:
    print(f"Error: '{csv_file}' not found.")
    print("Please upload the CSV file in the same folder as this notebook.")
```

## Dataset Information

```
if 'df' in globals():
    print(df.info())
    display(df.describe(include='all'))
```

## Missing Values

```
if 'df' in globals():
    display(df.isnull().sum())
```

## Data Visualization

```
if 'df' in globals():
    import matplotlib.pyplot as plt

    df.hist(figsize=(12, 8))
    plt.tight_layout()
    plt.show()
```
